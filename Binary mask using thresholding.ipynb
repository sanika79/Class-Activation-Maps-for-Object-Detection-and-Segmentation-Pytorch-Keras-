### creation of a binary mask for segmentation

```python
frame = cv2.imread(r"C:\Users\Sanika\Desktop\LAB Work\PennFudanPed\sample1.jpg", cv2.IMREAD_GRAYSCALE) 

print(frame.dtype)
print(frame.shape) 

_, contours, _ = cv2.findContours(frame, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

#cv2.imshow('Image' , frame)
#cv2.waitKey()

mask = np.zeros((frame.shape[0], frame.shape[1]), dtype=np.uint8)

print(mask.dtype)
print(mask.shape) 

#cv2.imshow('Mask' , mask)
#cv2.waitKey()
```

```python
new_frame = np.zeros(frame.shape, np.uint8)

cv2.imshow('New_frame' , new_frame)
cv2.waitKey() 
```

```python
for i, contour in enumerate(contours):
    c_area = cv2.contourArea(contour)
    if c_area > 6000:
        #mask = np.zeros(frame.shape, np.uint8)
        mask = cv2.drawContours(mask, contours, i, 255, cv2.FILLED)
        mask = cv2.bitwise_and(frame, mask) 
        new_frame = cv2.bitwise_or(new_frame, mask)
        frame = new_frame 

        cv2.imshow('Frame2', frame)
        cv2.waitKey() 

        #if self.contour_disp_flag:
            #frame = cv2.cvtColor(frame, cv2.COLOR_GRAY2BGR)
            #cv2.drawContours(frame, contours, -1, (255, 0, 0), 1)

```
